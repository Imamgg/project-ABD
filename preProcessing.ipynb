## 2. Data Loading

**Tujuan**: Memuat data mentah dari file CSV dengan struktur header yang benar.

````python
data_dir = Path('data')
print("Loading datasets with proper headers...\n")

def load_csv_with_headers(filepath):
    df_temp = pd.read_csv(filepath, nrows=4, header=None)
    
    product_names = df_temp.iloc[2].tolist()
    
    df = pd.read_csv(filepath, skiprows=4, header=None, names=product_names)
    
    return df

# Buah-buahan
buah_2023_raw = load_csv_with_headers(data_dir / 'Rata-rata Pengeluaran Perkapita Seminggu  Menurut Kelompok Buah-Buahan Per Kabupaten_kota, 2023.csv')
buah_2024_raw = load_csv_with_headers(data_dir / 'Rata-rata Pengeluaran Perkapita Seminggu  Menurut Kelompok Buah-Buahan Per Kabupaten_kota, 2024.csv')

# Sayur-sayuran
sayur_2023_raw = load_csv_with_headers(data_dir / 'Rata-rata Pengeluaran Perkapita Seminggu  Menurut Kelompok Sayur-Sayuran Per Kabupaten_kota, 2023.csv')
sayur_2024_raw = load_csv_with_headers(data_dir / 'Rata-rata Pengeluaran Perkapita Seminggu  Menurut Kelompok Sayur-Sayuran Per Kabupaten_kota, 2024.csv')

print(f"Buah 2023: {buah_2023_raw.shape}")
print(f"  Columns: {list(buah_2023_raw.columns[:5])}...")
print(f"\nBuah 2024: {buah_2024_raw.shape}")
print(f"  Columns: {list(buah_2024_raw.columns[:5])}...")
print(f"\nSayur 2023: {sayur_2023_raw.shape}")
print(f"  Columns: {list(sayur_2023_raw.columns[:5])}...")
print(f"\nSayur 2024: {sayur_2024_raw.shape}")
print(f"  Columns: {list(sayur_2024_raw.columns[:5])}...")

print("\nData loaded successfully!")
````

### 2.1 Custom CSV Loading Function

**Penjelasan Teknis**:

Fungsi `load_csv_with_headers()` dirancang khusus untuk menangani struktur CSV yang kompleks:

**Langkah-langkah**:
1. **Baca header** (4 baris pertama) untuk mendapatkan nama kolom dari baris ke-2
2. **Extract product names** dari baris index 2
3. **Load data actual** mulai dari baris ke-5 (index 4) dengan nama kolom yang sudah didapat

**Mengapa perlu custom function**:
- Fungsi `pd.read_csv()` biasa tidak bisa handle multi-level headers dengan baik
- Perlu kontrol manual untuk memilih baris mana yang jadi nama kolom
- Memastikan nama produk yang deskriptif (bukan '2023' atau 'nan')

**Parameter yang digunakan**:
- `nrows=4`: Hanya baca 4 baris pertama untuk ekstrak header
- `skiprows=4`: Lewati 4 baris pertama saat load data actual
- `header=None`: Jangan gunakan baris pertama sebagai header otomatis
- `names=product_names`: Gunakan nama kolom yang sudah kita ekstrak

## 3. Data Cleaning

**Tujuan**: Membersihkan data dari inkonsistensi, missing values, dan error.

### 3.1 Cleaning Function - Remove Header Rows and Fix Structure

**Penjelasan**:
Fungsi `clean_dataset()` melakukan:
1. **Rename kolom pertama**: Mengubah nama kolom pertama (yang bernama `nan`) menjadi 'Kabupaten_Kota'
2. **Hapus baris kosong**: Menghapus baris yang tidak memiliki nama kabupaten/kota
3. **Tambah metadata**: Menambahkan kolom 'Tahun' dan 'Kategori' untuk identifikasi

**Mengapa diperlukan**:
- Data mentah memiliki struktur yang tidak konsisten
- Perlu identifikasi jelas untuk setiap baris (kabupaten, tahun, kategori)
- Memudahkan proses analisis selanjutnya

**Output**: DataFrame yang bersih dengan struktur kolom yang konsisten.

````python
def clean_dataset(df, year, category):
    df_clean = df.copy()
    
    cols = list(df_clean.columns)
    cols[0] = 'Kabupaten_Kota'
    df_clean.columns = cols
    
    df_clean = df_clean[df_clean['Kabupaten_Kota'].notna()].copy()
    df_clean = df_clean[df_clean['Kabupaten_Kota'].astype(str).str.strip() != ''].copy()
    df_clean = df_clean[df_clean['Kabupaten_Kota'].astype(str).str.strip() != 'nan'].copy()
    
    df_clean['Tahun'] = year
    df_clean['Kategori'] = category
    
    print(f"Cleaned {category} {year}: {df_clean.shape}")
    print(f"  Columns: {list(df_clean.columns[:5])}...")
    
    return df_clean

print("Cleaning datasets...\n")
buah_2023 = clean_dataset(buah_2023_raw, 2023, 'Buah')
buah_2024 = clean_dataset(buah_2024_raw, 2024, 'Buah')
sayur_2023 = clean_dataset(sayur_2023_raw, 2023, 'Sayur')
sayur_2024 = clean_dataset(sayur_2024_raw, 2024, 'Sayur')

print("\nAll datasets cleaned!")
````

### 3.2 Handle Missing Values and Data Type Conversion

**Tujuan**: Menangani nilai yang hilang dan mengkonversi tipe data yang tepat.

**Masalah yang ditangani**:
1. **Simbol '-'**: Dalam data, tanda '-' menunjukkan tidak ada pengeluaran untuk produk tersebut
2. **String kosong**: Beberapa sel kosong perlu dikonversi ke nilai numerik
3. **Tipe data mixed**: Kolom yang seharusnya numerik masih bertipe object/string

**Strategi**:
- Ganti '-' dan string kosong dengan **0** (artinya: tidak ada pengeluaran)
- Konversi semua kolom produk menjadi **float64** (tipe numerik)
- Tetap pertahankan kolom Kabupaten_Kota, Tahun, Kategori sebagai string/integer

**Mengapa menggunakan 0 bukan NaN**:
- Nilai 0 lebih masuk akal secara bisnis (tidak ada pengeluaran)
- Memudahkan perhitungan statistik
- Tidak perlu handling khusus untuk missing values

**Output**: DataFrame dengan tipe data yang konsisten dan tidak ada missing values.

````python
def handle_missing_and_convert(df):
    df_processed = df.copy()
    
    non_numeric_cols = ['Kabupaten_Kota', 'Tahun', 'Kategori']
    numeric_cols = [col for col in df_processed.columns if col not in non_numeric_cols]
    
    for col in numeric_cols:
        df_processed[col] = df_processed[col].replace('-', '0')
        df_processed[col] = df_processed[col].replace('', '0')
        
        df_processed[col] = pd.to_numeric(df_processed[col], errors='coerce')
    
    df_processed[numeric_cols] = df_processed[numeric_cols].fillna(0)
    
    df_processed['Kabupaten_Kota'] = df_processed['Kabupaten_Kota'].astype(str).str.strip()
    
    return df_processed

print("Processing missing values and converting data types...\n")
buah_2023_processed = handle_missing_and_convert(buah_2023)
buah_2024_processed = handle_missing_and_convert(buah_2024)
sayur_2023_processed = handle_missing_and_convert(sayur_2023)
sayur_2024_processed = handle_missing_and_convert(sayur_2024)

print("Missing values after processing:")
print(f"Buah 2023: {buah_2023_processed.isnull().sum().sum()}")
print(f"Buah 2024: {buah_2024_processed.isnull().sum().sum()}")
print(f"Sayur 2023: {sayur_2023_processed.isnull().sum().sum()}")
print(f"Sayur 2024: {sayur_2024_processed.isnull().sum().sum()}")
print("\nMissing values handled and data types converted!")
````

## 4. Data Transformation

**Tujuan**: Mengubah format dan struktur data agar lebih sesuai untuk analisis.

### 4.1 Normalize Column Names

**Penjelasan**:
- **Membersihkan nama kolom**: Menghapus spasi ekstra
- **Mempertahankan readability**: Nama produk tetap jelas (contoh: "Jeruk, jeruk bali" bukan "jeruk_jeruk_bali")
- **Konsistensi**: Semua nama kolom mengikuti format yang sama

**Mengapa penting**:
- Nama kolom yang konsisten memudahkan akses data
- Menghindari error karena spasi atau karakter khusus
- Mempermudah dokumentasi dan maintenance code

**Catatan**: Kita mempertahankan nama produk yang deskriptif agar mudah dipahami oleh stakeholder non-teknis.

````python
def normalize_column_names(df):
    df_norm = df.copy()
    
    col_mapping = {}
    for col in df_norm.columns:
        if col not in ['Kabupaten_Kota', 'Tahun', 'Kategori']:
            new_col = col.strip()
            col_mapping[col] = new_col
    
    df_norm.rename(columns=col_mapping, inplace=True)
    
    return df_norm

buah_2023_norm = normalize_column_names(buah_2023_processed)
buah_2024_norm = normalize_column_names(buah_2024_processed)
sayur_2023_norm = normalize_column_names(sayur_2023_processed)
sayur_2024_norm = normalize_column_names(sayur_2024_processed)

print("Column names normalized!")
print(f"\nBuah columns: {list(buah_2023_norm.columns[:5])}...")
print(f"Sayur columns: {list(sayur_2023_norm.columns[:5])}...")
````

### 4.2 Create Long Format for Analysis (Melt Data)

**Tujuan**: Mengubah data dari format wide (lebar) menjadi long (panjang).

**Format Wide vs Long**:

**Wide Format** (sebelum):
```
Kabupaten | Apel | Mangga | Jeruk
Jakarta   | 100  | 200    | 150
Bandung   | 120  | 180    | 160
```

**Long Format** (sesudah):
```
Kabupaten | Produk | Pengeluaran
Jakarta   | Apel   | 100
Jakarta   | Mangga | 200
Jakarta   | Jeruk  | 150
Bandung   | Apel   | 120
...
```

**Keuntungan Long Format**:
1. âœ… Lebih fleksibel untuk filtering dan grouping
2. âœ… Ideal untuk visualisasi dengan library seperti Seaborn, Plotly
3. âœ… Sesuai dengan prinsip "tidy data"
4. âœ… Memudahkan analisis time-series
5. âœ… Lebih efisien untuk machine learning

**Kolom hasil**:
- `Kabupaten_Kota`: Nama wilayah
- `Tahun`: 2023 atau 2024
- `Kategori`: Buah atau Sayur
- `Produk`: Nama produk spesifik
- `Pengeluaran`: Nilai pengeluaran (Rupiah/Kapita/Minggu)
- `Is_Total`: Flag untuk kolom total kategori

**Output**: Dataset dalam format panjang yang siap untuk analisis mendalam.

````python
def transform_to_long_format(df):
    id_vars = ['Kabupaten_Kota', 'Tahun', 'Kategori']
    
    product_cols = [col for col in df.columns if col not in id_vars]
    total_col = product_cols[0]
    
    df_long = df.melt(
        id_vars=id_vars,
        value_vars=product_cols,
        var_name='Produk',
        value_name='Pengeluaran'
    )
    
    df_long['Is_Total'] = df_long['Produk'] == total_col
    
    return df_long

print("Transforming to long format...\n")
buah_2023_long = transform_to_long_format(buah_2023_norm)
buah_2024_long = transform_to_long_format(buah_2024_norm)
sayur_2023_long = transform_to_long_format(sayur_2023_norm)
sayur_2024_long = transform_to_long_format(sayur_2024_norm)

print(f"Buah 2023 long: {buah_2023_long.shape}")
print(f"Buah 2024 long: {buah_2024_long.shape}")
print(f"Sayur 2023 long: {sayur_2023_long.shape}")
print(f"Sayur 2024 long: {sayur_2024_long.shape}")

print("\nData transformed to long format!")
print("\nSample long format data:")
print(buah_2023_long.head(10))
````

### 5.2 Create Wide Format Integrated Dataset (for specific analyses)

**Tujuan**: Membuat dataset format wide yang menggabungkan Buah dan Sayur.

**Penjelasan**:
Meskipun format long lebih fleksibel, format wide tetap berguna untuk:
- Analisis korelasi antar produk
- Regresi dengan multiple features
- Pivot tables dan crosstabs
- Ekspor ke Excel untuk stakeholder

**Proses**:
1. Gabungkan data Buah 2023 & 2024
2. Gabungkan data Sayur 2023 & 2024
3. Tambahkan prefix pada nama kolom (Buah_ dan Sayur_)
4. Merge berdasarkan Kabupaten_Kota dan Tahun

**Hasil**:
- 1,028 baris (514 kabupaten Ã— 2 tahun)
- 47 kolom (metadata + semua produk Buah & Sayur)

**Use case**:
- Analisis: "Apakah kabupaten yang banyak beli buah juga banyak beli sayur?"
- Model prediksi dengan semua fitur sekaligus
- Dashboard dengan multiple metrics

**Output**: Dataset wide yang comprehensive dengan semua fitur.

````python
df_buah_combined = pd.concat([buah_2023_norm, buah_2024_norm], ignore_index=True)
df_sayur_combined = pd.concat([sayur_2023_norm, sayur_2024_norm], ignore_index=True)

print("Wide format combined datasets:")
print(f"Buah combined: {df_buah_combined.shape}")
print(f"Sayur combined: {df_sayur_combined.shape}")

df_buah_combined_renamed = df_buah_combined.copy()
df_sayur_combined_renamed = df_sayur_combined.copy()

for col in df_buah_combined_renamed.columns:
    if col not in ['Kabupaten_Kota', 'Tahun', 'Kategori']:
        df_buah_combined_renamed.rename(columns={col: f'Buah_{col}'}, inplace=True)

for col in df_sayur_combined_renamed.columns:
    if col not in ['Kabupaten_Kota', 'Tahun', 'Kategori']:
        df_sayur_combined_renamed.rename(columns={col: f'Sayur_{col}'}, inplace=True)

df_integrated_wide = df_buah_combined_renamed.merge(
    df_sayur_combined_renamed,
    on=['Kabupaten_Kota', 'Tahun'],
    how='outer',
    suffixes=('_Buah', '_Sayur')
)

print(f"\nFully integrated wide format: {df_integrated_wide.shape}")
print(f"\nColumns: {len(df_integrated_wide.columns)}")
````

âœ… Memperkaya context untuk machine learning
âœ… Memenuhi kebutuhan analisis bisnis

**Output**: Dataset yang diperkaya dengan 4 fitur tambahan yang bermakna.

````python
df_final_long = df_integrated_long.copy()

print("Adding derived features...\n")

df_final_long['Region'] = df_final_long['Kabupaten_Kota'].apply(
    lambda x: 'Aceh' if 'Aceh' in str(x) else 'Lainnya'
)

def categorize_expenditure(value):
    if value == 0:
        return 'Tidak Ada'
    elif value < 100:
        return 'Sangat Rendah'
    elif value < 500:
        return 'Rendah'
    elif value < 1000:
        return 'Sedang'
    elif value < 2000:
        return 'Tinggi'
    else:
        return 'Sangat Tinggi'

df_final_long['Kategori_Pengeluaran'] = df_final_long['Pengeluaran'].apply(categorize_expenditure)

df_final_long['Pengeluaran_Bulanan'] = df_final_long['Pengeluaran'] * 4

df_final_long['Pengeluaran_Tahunan'] = df_final_long['Pengeluaran'] * 52

print("Derived features added!")
print(f"\nNew columns: {list(df_final_long.columns)}")
print(f"\nCategories of expenditure:")
print(df_final_long['Kategori_Pengeluaran'].value_counts())
````
